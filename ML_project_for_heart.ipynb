```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
df= pd.read_csv('/content/heart_disease_data.csv')
```

```python
X = df.drop(columns='target', axis=1)
Y = df['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
model1=KNeighborsClassifier()
```

```python
model.fit(X_train.values, Y_train.values)
```

```python
model1.fit(X_train.values, Y_train.values)
```

```python
y_pred = model.predict(X_train.values)
training_data_accuracy = accuracy_score(y_pred, Y_train)
print('Accuracy on training data of Logistic :',training_data_accuracy)
```

```python
y_pred = model1.predict(X_train.values)
training_data_accuracy1 = accuracy_score(y_pred, Y_train)
print('Accuracy on training data KNN:',training_data_accuracy1)
```

```python
X_test_prediction = model.predict(X_test.values)
logistic_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data for logistic : ', logistic_accuracy)
```

```python
X_test_prediction = model1.predict(X_test.values)
Knn_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data for for knn : ', Knn_accuracy)
```

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

data= np.array(input_data)

input_data_reshaped = data.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```

```python
import matplotlib.pyplot as plt

plt.barh(['Logistic Regression', 'KNN'], [logistic_accuracy,Knn_accuracy ])


plt.title('Accuracy Comparison of Logistic Regression and KNN')
plt.xlabel('Model')
plt.ylabel('Accuracy')

plt.show()
```
